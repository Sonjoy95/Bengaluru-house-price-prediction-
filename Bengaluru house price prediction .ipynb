## Using Linear Regression algorithm

```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn.datasets as data
```

```python
df=pd.read_csv("Downloads/Cleaned_Bengaluru_House_Data.csv")
df
```

```python
x=df.drop("price",axis=1)
y=df["price"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=51)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

Above we have splitted the **input dataset (x)** into **x_train & x_test dataset** and **output dataset (y)** into **y_train & y_test dataset**.

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
```

We have applied the **Standardization** technique on **x_train & x_test** dataset using **StandardScaler** function.**Standardization** technique scales the dataset between **-3 to +3** and returns the 2D array. you can see below.

```python
sc=StandardScaler()
x_train_SSc=sc.fit_transform(x_train)
x_test_SSc=sc.fit_transform(x_test)
```

```python
x_train_SSc
```

```python
x_test_SSc
```

```python
from sklearn.linear_model import LinearRegression 
```

**Creating and applying linear regression model**

```python
lr=LinearRegression ()
lr.fit(x_train_SSc,y_train)
```

**regression coefficients ($\beta_1,\beta_2,\beta_3,.......$)**

```python
lr.coef_
```

```python
lr.intercept_
```

```python
x_test_SSc[0,:]
```

**Making Prediction**

```python
lr.predict([x_test_SSc[0,:]])
```

```python
pred=lr.predict(x_test_SSc)
pred
```

```python
y_test
```

**R2 score on test and train dataset**

```python
lr.score(x_test_SSc,y_test)
```

```python
lr.score(x_train_SSc,y_train)
```

```python
rd.coef_
```

```python
rd.intercept_
```

```python
rd.alpha
```

```python
rd.score(x_test_SSc,y_test)
```

```python
rd1=Ridge(alpha=2)
rd1.fit(x_train_SSc,y_train)
rd1.score(x_test_SSc,y_test)
```

## (ii) Lasso Regression

```python
ls=Lasso()
ls.fit(x_train_SSc,y_train)
```

```python
ls.coef_
```

```python
ls.intercept_
```

```python
ls.score(x_test_SSc,y_test)
```

```python
ls1=Lasso(alpha=2)
ls1.fit(x_train_SSc,y_train)
ls1.score(x_test_SSc,y_test)
```

In **Ridge & Lasso Regression** **alpha ($\alpha$)** is the parameter which balances the amount of emphasis given to minimizing RSS vs minimizing sum of square of coefficients.

i.e if $\alpha=0$:  The objective become same as linear regression, i.e. we will get the same ccoefficient as simple linear regression.

if **$\alpha=\infty$**:  The coefficients will be zero.

if **$0<\alpha<\infty$**: The magnitude of $\alpha$ will decide the weightage given to different parts of objective.

from above regression algorithms with regularization techniques **linear regression & Ridge regression** performs same based on **r2 score** but **Lasso regression** underperformed as campared to **Linear and Ridge**.

For more better performance lets use some more algorithms.

## Using Polynomial Regression

```python
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures
```

To apply **Ploynomial Regression** on a certain dataset, first we have to convert that dataset into polynomial. Using **PolynomialFeatures** function we can convert dataset into polynomial. You can see below the procedure.

```python
poly_reg=PolynomialFeatures(degree=2)
x_train_SSc_poly=poly_reg.fit_transform(x_train_SSc)
x_test_SSc_poly=poly_reg.fit_transform(x_test_SSc)
print(x_train_SSc_poly.shape,x_test_SSc_poly.shape)
```

**Applying Linear regression on plynomial dataset**

```python
poly_lr=LinearRegression ()
poly_lr.fit(x_train_SSc_poly,y_train)
```

```python
poly_lr.coef_
```

```python
poly_lr.intercept_
```

```python
y_predict1=poly_lr.predict(x_test_SSc_poly)
y_predict1
```

**r2 score**

```python
poly_lr.score(x_test_SSc_poly,y_test)
```

```python
y_test
```

As compared to ridge and linear **Polynomial** gives better performance.
